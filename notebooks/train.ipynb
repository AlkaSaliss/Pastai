```python
import os
import onnxruntime as onnx_rt
import numpy as np
from PIL import Image
import torch
import time
import torchvision
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import sys
sys.path.append("/data/Documents/Pastai/src/")
```

```python
from utils import non_max_suppression
from pastai_app import predict, post_process_img
```

```python
sess = onnx_rt.InferenceSession("../output/yolo_colab/exp3/weights/best.onnx")
```

```python
vf = cv2.VideoCapture("../inputs/data/videos/extract10s.mp4")
list_images = []
fps = 1
it = 0
while vf.isOpened():
    ret, img = vf.read()
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    
    if it%fps == 0:
        img = Image.fromarray(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        preds = predict(sess, img, 0.25, 0.45)
        img = np.array(img.resize((256, 256)))
        img = post_process_img(img, preds)
        list_images.append(img)
    it+=1
    
```
